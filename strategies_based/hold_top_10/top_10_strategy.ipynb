```python
import os
import pandas as pd

# Path to the folder containing CSV files
folder_path = 'C:\\Users\\Yash\\Data'

# Initialize an empty DataFrame
merged_df = pd.DataFrame()

# Loop through files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # Read the CSV file and extract 'Date' and 'Adj Close' columns
        df = pd.read_csv(file_path)
        ticker_symbol = filename[:-4]  # Extract the ticker symbol from the filename
        
        # Rename the 'Adj Close' column to the ticker symbol
        df.rename(columns={'Adj Close': ticker_symbol}, inplace=True)
        
        df = df.drop(['Open',  'High', 'Low', 'Close', 'Volume' ], axis=1)

        
        # Set 'Date' column as the index
        df.set_index('Date', inplace=True)
        
        # Merge the current DataFrame with the main DataFrame
        if merged_df.empty:
            merged_df = df
        else:
            merged_df = merged_df.merge(df, how='outer', left_index=True, right_index=True)

# Display the merged DataFrame
print(merged_df)
```

## price converted to percent change

```python
mtl = (merged_df.pct_change()+1)[1:].resample('M').prod()
```

```python
mtl
```

```python
mtl = mtl.fillna(1.000)
```

```python
mtl.replace([np.inf, -np.inf], 1.000, inplace=True)
```

```python
mtl
```

```python
def get_rolling_rate(df,n):
    return df.rolling(n).apply(np.prod)
```

# get the 12,6,3 months best 

```python
ret_12 , ret_6, ret_3 = get_rolling_rate(mtl,12),get_rolling_rate(mtl,6),get_rolling_rate(mtl,3)
```

```python
top_50 = ret_12.loc['2010-12-31'].nlargest(50).index
```

```python
top_30 = ret_6.loc['2010-12-31', top_50 ].nlargest(30).index
```

```python
top_10 = ret_3.loc['2010-12-31', top_30 ].nlargest(10).index
```

```python
top_10 , top_30, top_50
```

# select that top_10

```python
def get_top(date):
    top_50 = ret_12.loc[date].nlargest(50).index
    top_30 = ret_6.loc[date, top_50 ].nlargest(30).index
    top_10 = ret_3.loc[date, top_30 ].nlargest(10).index
    return top_10
```

```python
get_top('2022-12-31')
```

```python
portfolio = mtl.loc['2022-12-31':,get_top('2022-12-31')][1:2]
```

```python
portfolio.mean(axis=1)
```

```python
def pf_performance(date):
    portfolio = mtl.loc[date:,get_top(date)][1:2]
    return portfolio.mean(axis=1).values[0]
```

```python
mtl.index[:-1]
```

```python
returns = []

for date in mtl.index[:-1]:
    returns.append(pf_performance(date))
```
